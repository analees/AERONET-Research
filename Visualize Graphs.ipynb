```python
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
```

```python
#Read the NO2 data:
Thomas = pd.read_csv('data.csv')
Elena = pd.read_csv('data2.csv')
```

```python
Thomas.head()
```

```python
Elena.head()
```

```python
Thomas.columns = Thomas.columns.str.strip()
Elena.columns = Elena.columns.str.strip()
```

```python
Elena = Elena[(Elena['Data_quality_flag'] == 1) | (Elena['Data_quality_flag'] == 10)]
```

## Comparison of NO2 Values With Diffirente Scale

```python
import plotly.graph_objects as go
import plotly.io as pio

# Create traces
trace1 = go.Scatter(
    x=Thomas['Datetime'],
    y=Thomas['NO2_total_vertical_column'],
    mode='markers',
    name='Thomas Data'
)

trace2 = go.Scatter(
    x=Elena['Datetime'],
    y=Elena['NO2_total_vertical_column'],
    mode='markers',
    name='Elena Data',
    yaxis='y2'  # Assigning to the second y-axis
)

# Create the layout
layout = go.Layout(
    title='Comparison of NO2 Values',
    xaxis=dict(title='Datetime'),
    yaxis=dict(title='Dataset 1 NO2 Total Vertical Column'),
    yaxis2=dict(title='Dataset 2 NO2 Total Vertical Column', overlaying='y', side='right')
)

# Create the figure and add traces
data = [trace1, trace2]
fig = go.Figure(data=data, layout=layout)

# Show the plot
pio.show(fig)
```

## Comparison of NO2 Values With The Same Scale

```python
import plotly.graph_objects as go
import plotly.io as pio

# Create traces
trace1 = go.Scatter(
    x=Thomas['Datetime'],
    y=Thomas['NO2_total_vertical_column'],
    mode='markers',
    name='Thomas Data'
)

trace2 = go.Scatter(
    x=Elena['Datetime'],
    y=Elena['NO2_total_vertical_column'],
    mode='markers',
    name='Thomas Data',
    yaxis='y2'  # Assigning to the second y-axis
)

# Find the maximum value of both datasets
max_value = max(Thomas['NO2_total_vertical_column'].max(), Elena['NO2_total_vertical_column'].max())

# Create the layout
layout = go.Layout(
    title='Comparison of NO2 Values',
    xaxis=dict(title='Datetime'),
    yaxis=dict(title='NO2 Total Vertical Column', range=[0, max_value]),
    yaxis2=dict(title='NO2 Total Vertical Column', overlaying='y', side='right', range=[0, max_value])
)

# Create the figure and add traces
data = [trace1, trace2]
fig = go.Figure(data=data, layout=layout)

# Show the plot
pio.show(fig)
```
