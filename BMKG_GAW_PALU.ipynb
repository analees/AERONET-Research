```python
BMKG_GAW_PALU_col_compare = ['AOD_340nm-Total', 'AOD_380nm-Total', 'AOD_440nm-Total', 'AOD_500nm-Total', 'AOD_675nm-Total', 'AOD_870nm-Total', 'AOD_1020nm-Total', 'AOD_1640nm-Total']


# extract the values from the separate datasets which were defined in the prior cell
values_2024 = year_2024_data[BMKG_GAW_PALU_col_compare]
values_2023 = year_2023_data[BMKG_GAW_PALU_col_compare]

# Create subplots for side-by-side histograms
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

# Compare the distributions of the comparable values using histograms
for i, (values, dataset) in enumerate(zip([values_2024, values_2023], ['year_2024_data', 'year_2023_data'])):
    ax = axes[i]
    for col in BMKG_GAW_PALU_col_compare:
        ax.hist(values[col], bins=20, alpha=0.5, label=col)
    ax.set_xlabel('AOD Values')
    ax.set_ylabel('Frequency')
    ax.set_title(f'Comparison of AOD Values between {dataset}')
    ax.legend()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Convert AOD columns to numeric, drop negative values, and reset index
BMKG_GAW_PALU_col = ['AOD_340nm-Total', 'AOD_380nm-Total', 'AOD_440nm-Total', 'AOD_500nm-Total',
                     'AOD_675nm-Total', 'AOD_870nm-Total', 'AOD_1020nm-Total', 'AOD_1640nm-Total']
for col in BMKG_GAW_PALU_col:
    BMKG_GAW_PALU[col] = pd.to_numeric(BMKG_GAW_PALU[col], errors='coerce')
    BMKG_GAW_PALU = BMKG_GAW_PALU.drop(BMKG_GAW_PALU[BMKG_GAW_PALU[col] < 0].index)
BMKG_GAW_PALU = BMKG_GAW_PALU.reset_index(drop=True)

# Convert 'Date(dd:mm:yyyy)' to datetime object and extract year, month, and date
BMKG_GAW_PALU['Date(dd:mm:yyyy)'] = pd.to_datetime(BMKG_GAW_PALU['Date(dd:mm:yyyy)'], format='%d:%m:%Y')
BMKG_GAW_PALU['Year'] = BMKG_GAW_PALU['Date(dd:mm:yyyy)'].dt.year
BMKG_GAW_PALU['Month'] = BMKG_GAW_PALU['Date(dd:mm:yyyy)'].dt.month
BMKG_GAW_PALU['Date'] = BMKG_GAW_PALU['Date(dd:mm:yyyy)'].dt.date

# Filter for the year 2024
year_2024_data = BMKG_GAW_PALU[BMKG_GAW_PALU['Year'] == 2024]

# Convert start_date and end_date to datetime.date objects
start_date = pd.to_datetime('2024-05-05').date()
end_date = pd.to_datetime('2024-05-09').date()

# Filter data for the specified date range
day_2024_data = year_2024_data[(year_2024_data['Date'] >= start_date) & (year_2024_data['Date'] <= end_date)]

# Create 'Date' column within year_2024_data
year_2024_data['Date'] = pd.to_datetime(year_2024_data['Date(dd:mm:yyyy)'], format='%Y-%m-%d').dt.date

# Create the line plot
plt.figure(figsize=(10, 6))
for col in BMKG_GAW_PALU_col:
    plt.plot(day_2024_data['Date'], day_2024_data[col], label=col)  # Plot data within the date range as lines

plt.title('Indonesia AOD by Date')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()

```

```python
#list of columns to be observed as well as a for loop to convert the columns to numeric and reset the indeces
BMKG_GAW_PALU_col = ['AOD_340nm-Total', 'AOD_380nm-Total', 'AOD_440nm-Total', 'AOD_500nm-Total',
                     'AOD_675nm-Total', 'AOD_870nm-Total', 'AOD_1020nm-Total', 'AOD_1640nm-Total']
for col in BMKG_GAW_PALU_col:
    BMKG_GAW_PALU[col] = pd.to_numeric(BMKG_GAW_PALU[col], errors='coerce')
    BMKG_GAW_PALU = BMKG_GAW_PALU.drop(BMKG_GAW_PALU[BMKG_GAW_PALU[col] < 0].index)
BMKG_GAW_PALU = BMKG_GAW_PALU.reset_index(drop=True)

# Convert 'Date(dd:mm:yyyy)' to datetime object and extract year, month, and date
BMKG_GAW_PALU['Date(dd:mm:yyyy)'] = pd.to_datetime(BMKG_GAW_PALU['Date(dd:mm:yyyy)'], format='%d:%m:%Y')
BMKG_GAW_PALU['Year'] = BMKG_GAW_PALU['Date(dd:mm:yyyy)'].dt.year
BMKG_GAW_PALU['Month'] = BMKG_GAW_PALU['Date(dd:mm:yyyy)'].dt.month
BMKG_GAW_PALU['Date'] = BMKG_GAW_PALU['Date(dd:mm:yyyy)'].dt.date

# gather the 2024 data column
year_2024_data = BMKG_GAW_PALU[BMKG_GAW_PALU['Year'] == 2024]

# collected the desired months for the year 2024(April, May, June)
year_2024_data['Date'] = pd.to_datetime(year_2024_data['Date(dd:mm:yyyy)'], format='%Y-%m-%d').dt.date
april_readings = year_2024_data[year_2024_data['Month'] == 4].sort_values(by='Date')
may_readings = year_2024_data[year_2024_data['Month'] == 5].sort_values(by='Date')
june_readings = year_2024_data[year_2024_data['Month'] == 6].sort_values(by='Date')

# Create the line plot for April readings
plt.figure(figsize=(12, 6))
for col in BMKG_GAW_PALU_col:
    plt.plot(april_readings['Date'], april_readings[col],".",label=col)

plt.title('BMKG_GAW_PALU April Readings')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()

# Create the line plot for May readings
plt.figure(figsize=(12, 6))
for col in BMKG_GAW_PALU_col:
    plt.plot(may_readings['Date'], may_readings[col],".", label=col)

plt.title('BMKG_GAW_PALU May Readings')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()

# Create the line plot for June readings
plt.figure(figsize=(20, 6))
for col in BMKG_GAW_PALU_col:
    plt.plot(june_readings['Date'], june_readings[col],".", label=col)

plt.title('BMKG_GAW_PALU June Readings')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()


```
